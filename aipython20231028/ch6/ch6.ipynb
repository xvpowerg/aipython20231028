## 輸入四個月的支出金額後列出最多與最少的支出金額

```python
priceList=[]
priceSum = 0
for i in range(1,5):
    p = int(input(f"請輸入{i}個月的支出金額"))
    priceSum =priceSum + p
    priceList.append(p)
priceList.sort()
print(f"max:{priceList[len(priceList)-1]}") 
print(f"min:{priceList[0]}")
print(priceSum)    
print(priceList) 

print(sum(priceList)) 
print(max(priceList))    
print(min(priceList))
```

## 使用raise拋例外

```python
def testScore(score):
    if score < 0 or score > 100:
        #print("成績錯誤")
        raise OverflowError("成績錯誤")
    return score
print(testScore(65))
try:
    print(testScore(-2))
except OverflowError as er :
    print("OverflowError:",er)
```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
import re
text = 'one, two...ten'
re1 = re.split("[,. ]+",text)
re2 = re.split("[,. ]+",text,maxsplit=1)
print(re1)
print(re2)
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456"
rStr1 = re.sub("\d+","222",inputStr)
rStr2 = re.sub("\d+","222",inputStr,count=1)
print(rStr1)
print(rStr2)
```

## 繼承

```python
class Animal:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def printInfo(self):
        print(self.name,self.age)
a1 =  Animal("Bobo",5)
a1.printInfo()
```

```python
class Cat(Animal):
    pass
cat1 = Cat("Kitty",10)
cat1.printInfo()
```

```python
class Dog(Animal):
    def __init__(self,name,age,weight):
        super().__init__(name,age)#呼叫父類別的__init__
        self.weight = weight
    def printInfo(self):
        super().printInfo()
        print(self.weight)
dog1 = Dog("Bobo",5,10)
dog1.printInfo()
```

## 多重繼承與初始化方法
### 多重繼承
+ 繼承多個父類別時用，隔開
+ 注意成員搜尋的順序
  + 從子類別開始
  + 同一階層父類別由左至右搜尋
  + 更上層祖父類別由左至右搜尋
  + 直到達到頂層為止

```python
class S:
    def method1(self):
        print("S.Method1")
    def method2(self):
        print("S.Method2")
class A(S):
    def method3(self):
         print("A.Method3")
class B(S):
    def method2(self):
        print("B.Method2")
    def method3(self):
        print("B.Method3")
class C(A,B):
    def method4(self):
        print("C.Method4")
c = C()
c.method4()
c.method3()
c.method2()
c.method1()
```

## __str__ 可輸出物件的內容

```python
class Point:
    def __init__(self,x = 0 ,y = 0):
        self.x = x
        self.y = y
p1 = Point(2,3)
p2 = Point(-1,2)

print(p1)
print(p2)
```

```python
class Point:
    def __init__(self,x = 0 ,y = 0):
        self.x = x
        self.y = y
    def __str__(self):
        return f"({self.x},{self.y})"
p1 = Point(2,3)
p2 = Point(-1,2)

print(p1)
print(p2)
```
