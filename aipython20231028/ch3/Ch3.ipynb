## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "Python"
print(str1.find(search1))
search2 = "not"
print(str1.find(search2))
search3 = "t"
print(str1.find(search3))
search4 = "t"
print(str1.find(search4,11))#第11位置開始搜尋
```

# List VS Tuple

+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 可以隨時修改

## list 介紹
+ list 內資料用逗號隔開，放置於中括號的範圍內。
  + list_name = [1,2,3 ]
+ list 可以儲存不同類型的值或項目，如數字和字串。
  + list_name = [1, 2, ‘Python’, ‘lists’ ] 
+ list 內的資料可以進行增加、更新與移除。
+ 可透過.index( )方法取得指定資料的索引。

```python
list1 = ["H",'i']
tuple1 = ('H','i')
print(list1)
print(len(list1))
list1.append("Vivin")
print(list1)
print(len(list1))
list1[1] = 3.1415
print(list1)
print("=======================")
print(tuple1)
#tuple1[0] = "Lucy" #不可改只能讀
print(tuple1[1])
```

```python
list2 = [1,2,3,4]
list2.append(5)
list2.append(6)
result = tuple(list2)#list to tuple
print(result)
print(type(result))
list2 = list(result)#tuple to list
print(list2)
print(type(list2))
```

## list 項目擴展資料
+ extend 擴展資料：
  + 如果引號括起來的字串有 [ ] 將會拆解成好幾個資料。
  + 可以把其他 list 加入到這個 list 內，擴展多個欄位。

```python
list1 = [1,2,3,4,5]
list2 = [21,22,23,24,25]
print(list1)
print(id(list1))
list1.extend(list2)
print(list1)
print(len(list1))
print(id(list1))
```

```python
list1 = [1,2,3,4,5]
list2 = [21,22,23,24,25]
print(id(list1))
list1 = list1 + list2
print(list1)
print(id(list1))
```

```python
list1 = [1,2,3,4,5]
list2 = [21,22,23,24,25]
print(id(list1))
print(list1)
list1 += list2
print(list1)
print(id(list1))
```

## list 項目插入新資料
+ list 可使用 insert 方式插入新的資料：
   + list.insert(索引值，插入的資料)。
   + 若索引值超出範圍則會改為 append 方式附加。

```python
list1 = ['a','b','c','d','e']
print(len(list1))
list1.insert(1,"x")
print(list1)
print(len(list1))
list1.insert(-2,"y")
print(list1)
list1.insert(-4,"z")
print(list1)
list1.insert(88,"g")
print(list1)
```

## set會過濾重複資料
