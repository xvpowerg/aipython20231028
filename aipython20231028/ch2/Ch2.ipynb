## FOR迴圈
### 知道數量時使用for loop
### python的for 有一種由頭到取數值的想法



```python
myRange = range(10)
print(type(myRange))
print(myRange)
print(list(myRange))
```

```python
myRange2 = range(1,10)
print(list(myRange2))
```

```python
myRange3 = range(1,10,2)
print(list(myRange3))
```

```python
myRange4 = range(10,1,-2)
print(list(myRange4))
```

```python
for i in range(5):
    print(i)
```

```python
for k in range(2,8):
    print(k,end=" ")
```

```python
for x in range(2,18,3):
    print(x)
```

```python
for y in range(10,1,-2):
    print(y)
```

```python
for i in range(4):
    print(i)
print("離開後的i:",i)    

for i in range(-4):
    print(i)
print("離開後的i:",i)     
```

```python
a = 6
while a > 0:
    print(a)
    a -= 1
```

```python
for i in range(1,6):
    for k in range(1,4):
        print(f"{i}-{k}",end=" ")
    print()    
```

## break 代表不回頭，跳出迴圈。
## continue 代表繼續執行下一個迴圈。


```python
for i in range(1,11):
    if i == 5:
        break
    print(i)
print("End")    
```

```python
for i in range(1,11):
    if i == 5 or i == 7:
        continue
    print(i)
print("End")    
```
